```python
#1. List Your Favorite Artists:
artists = ["Billie Eilish", "Absofacto", "Hermanos Gutiriez"]
```

```python
#2. Fetch Top Tracks:

#Write a function named get_top_tracks.
#This function should accept an artist's name and return the name of the first 5 top tracks by that artist.
#Use the function get_top_tracks to get the first 5 top tracks for each artist in your artists list.
#And store the results in a new list named top_tracks_list.

def get_access_token():
    url = "https://accounts.spotify.com/api/token"
    headers = {"Content-Type": "application/x-www-form-urlencoded"}
    data = {"grant_type": "client_credentials"}
    
    response = requests.post(url, headers=headers, data=data, auth=(CLIENT_ID, CLIENT_SECRET))
    response_data = response.json()
    
    return response_data.get("access_token")

# Function to get top 5 tracks of an artist
def get_top_tracks(artist_name):
    access_token = get_access_token()
    
    # Search for artist ID
    search_url = f"https://api.spotify.com/v1/search?q={artist_name}&type=artist"
    headers = {"Authorization": f"Bearer {access_token}"}
    
    response = requests.get(search_url, headers=headers)
    search_results = response.json()
    
    if not search_results["artists"]["items"]:
        return f"Artist '{artist_name}' not found."

    artist_id = search_results["artists"]["items"][0]["id"]

    # Get top tracks
    top_tracks_url = f"https://api.spotify.com/v1/artists/{artist_id}/top-tracks?market=US"
    response = requests.get(top_tracks_url, headers=headers)
    top_tracks_data = response.json()

    # Extract track names
    top_tracks = [track["name"] for track in top_tracks_data["tracks"][:5]]
    
    return top_tracks

top_tracks_list = {artist: get_top_tracks(artist) for artist in artists}

# Print results
for artist, tracks in top_tracks_list.items():
    print(f"\nTop tracks for {artist}:")
    for track in tracks:
        print(f"- {track}")
```

```python
#before finding related artists, we need to get the artist ID 

import requests
import urllib.parse  # For proper URL encoding

CLIENT_ID = "c3d9e7712a674604aae955ceeead682b"
CLIENT_SECRET = "e32ab9f5ca624c419e00c7a3ca4d0150"

# Function to get the access token
def get_access_token():
    url = "https://accounts.spotify.com/api/token"
    headers = {"Content-Type": "application/x-www-form-urlencoded"}
    data = {"grant_type": "client_credentials"}
    
    response = requests.post(url, headers=headers, data=data, auth=(CLIENT_ID, CLIENT_SECRET))
    
    if response.status_code != 200:
        print(f" Error: Failed to get access token (Status Code {response.status_code})")
        print(response.json())  # Print the full error message
        return None

    return response.json().get("access_token")

# Function to get the artist ID
def get_artist_id(artist_name):
    access_token = get_access_token()
    if not access_token:
        return " Error: Unable to authenticate with Spotify API."

    artist_name = artist_name.strip()  # Remove extra spaces
    encoded_artist_name = urllib.parse.quote(artist_name)  # Encode name for URL

    # Search for artist ID
    search_url = f"https://api.spotify.com/v1/search?q={encoded_artist_name}&type=artist"
    headers = {"Authorization": f"Bearer {access_token}"}
    
    response = requests.get(search_url, headers=headers)

    if response.status_code != 200:
        print(f" Error: Failed to search for '{artist_name}' (Status Code {response.status_code})")
        print(response.json())  # Print the full response for debugging
        return None

    search_results = response.json()

    # Check if "artists" key exists and contains results
    if "artists" not in search_results or not search_results["artists"]["items"]:
        return f" Artist '{artist_name}' not found in Spotify database."

    # Get the first artist's ID
    artist_id = search_results["artists"]["items"][0]["id"]
    return artist_id

# Example list of artists
artists = ["Billie Eilish", "Absofacto", "Hermanos Gutirez"]

# Get artist IDs for each artist
artist_ids = {artist: get_artist_id(artist) for artist in artists}

# Print results
for artist, artist_id in artist_ids.items():
    print(f" Artist: {artist} - Spotify ID: {artist_id}")
```

```python
import time
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
from spotipy.exceptions import SpotifyException

# Spotify API credentials
CLIENT_ID = "d657f7c2f38848b5b883e01043f091c1"
CLIENT_SECRET = "160fb8c6303e4b6c9bf822fbabc637d9"

# Authenticate using Spotipy
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=CLIENT_ID, client_secret=CLIENT_SECRET))

# Function to get an artist's ID
def get_artist_id(artist_name):
    result = sp.search(q=artist_name, type="artist", limit=1)
    if result and result["artists"]["items"]:
        artist_id = result["artists"]["items"][0]["id"]
        print(f" Found Artist ID for {artist_name}: {artist_id}")  # Debugging
        return artist_id
    else:
        print(f" Artist '{artist_name}' not found on Spotify.")
        return None

# Function to get related artists
def find_related_artists(artist_name):
    artist_id = get_artist_id(artist_name)
    if not artist_id:
        return f" No related artists found for '{artist_name}'."

    try:
        time.sleep(1)  # Small delay to avoid rate limiting
        related_artists = sp.artist_related_artists(artist_id)
        if "artists" not in related_artists or not related_artists["artists"]:
            return f" No related artists found for '{artist_name}'."

        return [artist["name"] for artist in related_artists["artists"][:5]]

    except SpotifyException as e:
        if e.http_status == 404:
            return f" No related artists data available for '{artist_name}' (Spotify API 404)."
        else:
            return f" Error fetching related artists for '{artist_name}': {e}"

# Test with my artists list:
artists = ["Billie Eilish", "Absofacto", "Hermanos Gutirez"]
for artist in artists:
    print(f"\n Related artists for {artist}:")
    print(find_related_artists(artist))
```

### Getting Artists from a Playlist

To extract all the artists from the tracks in a playlist, you'd typically follow these steps:

1. Fetch the playlist's tracks.
2. Iterate through each track.
3. For each track, extract the associated artists.

```python
# List to store unique artist names
artists_list = []

for track_item in tracks['items']:
    track = track_item['track']
    for artist in track['artists']:
        artist_name = artist['name']
        if artist_name not in artists_list:  # This ensures each artist is added only once
            artists_list.append(artist_name)

print(artists_list)
```

```python
#3. Track-tastic:
#Extract and display the names of the first 10 tracks in the chosen playlist


# Function to fetch featured playlists
def get_featured_playlists():
    try:
        playlists = sp.featured_playlists(limit=5)  # Fetch top 5 featured playlists
        return [(p["name"], p["id"]) for p in playlists["playlists"]["items"]]
    except Exception as e:
        return f" Error fetching featured playlists: {e}"

# Function to fetch details of a selected playlist
def get_playlist_details(playlist_id):
    try:
        playlist = sp.playlist(playlist_id)
        return {
            "name": playlist["name"],
            "description": playlist["description"],
            "track_count": playlist["tracks"]["total"],
        }
    except Exception as e:
        return f" Error fetching playlist details: {e}"

# Function to fetch the first 10 track names from a playlist
def get_playlist_tracks(playlist_id, limit=10):
    try:
        tracks = sp.playlist_tracks(playlist_id, limit=limit)["items"]
        return [track["track"]["name"] for track in tracks]
    except Exception as e:
        return f" Error fetching playlist tracks: {e}"

# Fetch top 5 featured playlists
featured_playlists = get_featured_playlists()

if isinstance(featured_playlists, list):
    # Choose a random playlist
    selected_playlist = random.choice(featured_playlists)
    selected_name, selected_id = selected_playlist

    # Fetch playlist details
    details = get_playlist_details(selected_id)
    track_names = get_playlist_tracks(selected_id, limit=10)

    # Print results
    print(f"\n **Selected Playlist:** {details['name']}")
    print(f" **Description:** {details['description']}")
    print(f" **Total Tracks:** {details['track_count']}\n")

    print(" **First 10 Tracks:**")
    for i, track in enumerate(track_names, 1):
        print(f"{i}. {track}")

else:
    print(featured_playlists)
```

```python
#4. Artistic Flair:
#Create a dictionary where the keys are the names of the first 10 tracks, and the values are lists containing the names of the artists associated with each track.

# Function to fetch featured playlists
def get_featured_playlists():
    try:
        playlists = sp.featured_playlists(limit=5)  # Fetch top 5 featured playlists
        return [(p["name"], p["id"]) for p in playlists["playlists"]["items"]]
    except Exception as e:
        return f" Error fetching featured playlists: {e}"

# Function to fetch the first 10 track names and their artists from a playlist
def get_playlist_tracks(playlist_id, limit=10):
    try:
        tracks = sp.playlist_tracks(playlist_id, limit=limit)["items"]

        # Create a dictionary {Track Name: [Artist1, Artist2, ...]}
        track_artist_dict = {
            track["track"]["name"]: [artist["name"] for artist in track["track"]["artists"]]
            for track in tracks
        }
        return track_artist_dict

    except Exception as e:
        return f" Error fetching playlist tracks: {e}"

# Fetch top 5 featured playlists
featured_playlists = get_featured_playlists()

if isinstance(featured_playlists, list):
    # Choose a random playlist
    selected_playlist = random.choice(featured_playlists)
    selected_name, selected_id = selected_playlist

    # Fetch track and artist details
    track_artist_dict = get_playlist_tracks(selected_id, limit=10)

    # Print results
    print(f"\n **Selected Playlist:** {selected_name}")
    print("\n **First 10 Tracks & Their Artists:**")
    for track, artists in track_artist_dict.items():
        print(f" {track}: {', '.join(artists)}")

else:
    print(featured_playlists)
```
